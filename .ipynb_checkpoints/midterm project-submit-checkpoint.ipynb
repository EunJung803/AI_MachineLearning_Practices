```python
from __future__ import print_function
import time
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

import torchvision
import torch
from torchvision import datasets
from torchvision import transforms

%matplotlib inline
```

```python
# A DNN with hyper parameters

import torch.nn as nn
import torch.optim as optim


class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.in_dim = 28*28 #MNIST
        self.out_dim = 10
        
        self.fc1 = nn.Linear(self.in_dim, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 64)
        self.fc5 = nn.Linear(64, self.out_dim)
        
        self.relu = nn.ReLU()
        self.log_softmax = nn.LogSoftmax()
        
    def forward(self, x):
        z1 = x.view(-1, self.in_dim)
        a1 = self.relu(self.fc1(z1))
        z2 = self.relu(self.fc2(a1))
        a2 = self.relu(self.fc3(z2))
        z3 = self.relu(self.fc4(a2))
        logit = self.fc5(z3)
        return logit, z1, a1, z2, a2
```

```python
model = MLP()
criterion = nn.CrossEntropyLoss() 
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
batch_size = 12

train_data = datasets.MNIST('D:\datasets', train=True, download=True, transform = transforms.ToTensor())
test_data = datasets.MNIST('D:\datasets', train=False, download=True, transform = transforms.ToTensor())

train_loader = torch.utils.data.DataLoader(train_data, batch_size = batch_size, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size = batch_size)


for epoch in range(10):  #loop over the dataset multiple times
    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):
        #get the inputs; data is a list of [inputs, labels]
        inputs, labels = data
        
        #zero the parameter gradients
        optimizer.zero_grad()
        
        #forward + backward + optimize
        outputs, z1, a1, z2, a2  = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        #print statistics
        running_loss += loss.item()
        if (i+1) % 2000 == 0:
            print('[%d, %5d] loss: %.3f' %
                    (epoch + 1, i + 1, running_loss / 2000))
            running_loss = 0.0
            
            
print('Finished Training')
```

```python
z1, a1, z2, a2


z1 = z1.numpy()

a1 = a1.detach().numpy()

z2 = z2.detach().numpy()

a2 = a2.detach().numpy()

outputs = outputs.detach().numpy()
```

```python
feat_cols_z1 = [f'data{i}' for i in range(z1.shape[1])]
z1_df = pd.DataFrame(z1, columns = feat_cols_z1)
z1_df['y'] = batch_size
z1_df
```

```python
feat_cols_a1 = [f'data{i}' for i in range(a1.shape[1])]
a1_df = pd.DataFrame(a1, columns = feat_cols_a1)
a1_df['y'] = batch_size
a1_df
```

```python
feat_cols_z2 = [f'data{i}' for i in range(z2.shape[1])]
z2_df = pd.DataFrame(z2, columns = feat_cols_z2)
z2_df['y'] = batch_size
z2_df
```

```python
feat_cols_a2 = [f'data{i}' for i in range(a2.shape[1])]
a2_df = pd.DataFrame(a2, columns = feat_cols_a2)
a2_df['y'] = batch_size
a2_df
```

```python
#z1 pca

pca = PCA(n_components=2)
pca_result = pca.fit_transform(z1_df[feat_cols_z1].values)
z1_df['pca-one'] = pca_result[:,0]
z1_df['pca-two'] = pca_result[:,1]
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))

plt.figure(figsize = (10,7))
sns.scatterplot(
    x = "pca-one", y = "pca-two",
    hue = "y",
    palette = sns.color_palette("hls", True),
    data = z1_df,
    legend = "full",
    alpha = 0.3
)
plt.show()
```

```python
#z1 tsne

rndperm = np.random.permutation(z1_df.shape[0])
N = 10000
df_subset = z1_df.loc[rndperm[:N],:].copy()
data_subset = df_subset[feat_cols_z1].values
pca = PCA(n_components = 2)
pca_result = pca.fit_transform(data_subset)
df_subset['pca-one'] = pca_result[:,0]
df_subset['pca-two'] = pca_result[:,1]
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))

time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(data_subset)
df_subset['tsne-2d-one'] = tsne_results[:,0]
df_subset['tsne-2d-two'] = tsne_results[:,1]
print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))

plt.figure(figsize=(10,7))
sns.scatterplot(
    x = "tsne-2d-one", y = "tsne-2d-two",
    hue = "y",
    palette = sns.color_palette("hls", True),
    data = df_subset,
    legend = "full",
    alpha = 0.3
)
plt.show()
```
