```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cluster2.csv')
print("Dimensions of the data = {}".format(df.shape))
```

```python
df[:5]
```

```python
X = df.values
X[:5]
```

```python
#set the size of the figure
plt.figure(figsize = (5, 5))

#plot the data points
plt.scatter(X[:, 0], X[:, 1], s=4)

#print the figure
plt.show()
```

```python
cmap = "tab10"
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=4, random_state=0)

y_pred = k_means.fit_predict(X)
print(y_pred[:10])
```

```python
print(k_means.cluster_centers_)
```

```python
plt.figure(figsize = (5, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=4, cmap=cmap)
#'c' : The color of each point

plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan= DBSCAN(eps=0.05, min_samples=20)
y_pred = dbscan.fit_predict(X)
print(y_pred[:10])
```

```python
for i, (eps, min_samples) in enumerate(
    ((0.05, 20), (0.06, 20), (0.06,15), (0.06,6))):
    dbscan=DBSCAN(eps=eps, min_samples=min_samples)
    y_pred = dbscan.fit_predict(X)
    plt.figure(i+1, figsize=(5,5))
    plt.scatter(X[:, 0], X[:, 1],
               cmap=cmap, s=4, c=y_pred)
    plt.title("eps: {}, min_samples: {}".format(eps, min_samples))
plt.show()
```
